```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
file_path = '/Users/christinemwanza/Downloads/diabetes.csv'  
data = pd.read_csv(file_path)
row_count= data.shape[0] #checking the number of rows the dataset contains
print(row_count)
```

```python
# shorter way of checking the No of rows
print(data.shape[0])
```

```python
# Visualizations

# Distribution of Age
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Glucose Levels by Diabetes Outcome
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='Glucose', data=data)
plt.title('Glucose Levels by Diabetes Outcome')
plt.xlabel('Diabetes Outcome')
plt.ylabel('Glucose Level')
plt.show()
```

```python
pairplot = sns.pairplot(data, hue='Outcome')
pairplot.fig.suptitle('Pairplot of Features Colored by Diabetes Outcome')
plt.show()

# Save the pairplot as an image file
pairplot.savefig('pairplot.png')
```

```python
import plotly.offline as pyo
import plotly.graph_objs as go
# Interactive Scatter Plot
fig = px.scatter(data, x='Age', y='Glucose', color='Outcome', title='Age vs Glucose Levels')
fig.show()
pyo.plot(fig, filename='interactive_plot.html', auto_open=False)
```
